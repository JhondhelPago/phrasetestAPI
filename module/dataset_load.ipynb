```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
from features_xtrct import PhraseExtract, PhraseExtract1
```

```python
df = pd.read_csv('PhraseTest_Data.csv')
```

```python
df
```

```python
df.info()
```

```python
features_list = list()

for index, row in df.iterrows():
    print(f"row:{index}")
    Phrase = PhraseExtract1(question='What do you want to be when you grow up? How will you get there?', text=row['Question 1'])
    print(Phrase.FeatureList())
    features_list.append(Phrase.FeatureList())
    
```

```python
print(features_list)
```

```python
features_category = ['Word_Count', 'UniqueWordsRatio', 'AvgWordLen', 'NounCount', 'ADJ_Count', 'ADV_Count', 'PRONOUN_Count', 'VERB_Count', 'CohesiveDev_Count', 'ReadabilityScore', 'AvgSentLen', 'SentSimple', 'SentCompound', 'SentComplex', 'TopicRelevance']
new_df = pd.DataFrame(features_list, columns=features_category)
```

```python
new_df
```

```python
scaler = StandardScaler()

scaled_data = scaler.fit_transform(new_df)
```

```python
print(scaled_data)
```

```python
kmeans = KMeans(n_clusters=3)  # Choose 3 clusters

# Step 4: Train the model
kmeans.fit(scaled_data)
```

```python
new_df['Cluster'] = kmeans.labels_

# Step 6: Visualize the clusters (if data is 2D or can be reduced to 2D)
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=new_df['Cluster'])
plt.show()
```
